```python
import networkx as nx
```

```python
X= nx.Graph()
X.add_nodes_from(['a', 'b', 'c', 'd','e'])
X.add_edges_from([('a','b'), ('b','c'), ('c','d'), ('a','c'),('a','e'),('d','e')])
nx.draw(X, with_labels = True)
```

```python
list(X.neighbors('b'))
```

```python
for neighbor in X.neighbors('b'):
    print(neighbor)
```

```python
X.has_node('a')
```

```python
nx.is_tree(X)
```

```python
nx.is_connected(X)
```

```python
X.degree('a')
```

```python
D = nx.DiGraph()
D.add_edges_from([('1','2'), ('2','3'), ('1','4'), ('3','4')])
nx.draw(D, with_labels = True)
```

```python
X=nx.Graph()
X.add_edges_from([(1,2),(2,3),(1,3),(3,4)])
nx.draw(X,with_labels=True,node_size=500)
```

```python
dict(X.degree())
```

```python
nx.to_numpy_matrix(X)
```
